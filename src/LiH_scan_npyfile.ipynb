```python
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib import rcParams
from mpl_toolkits.mplot3d import Axes3D
```

```python
# define color palette
COLOUR1 = "firebrick"
COLOUR2 = "green"
COLOUR3 = "royalblue"
COLOUR4 = "rebeccapurple"
COLOUR5 = 'darkorchid'
COLOUR6 = 'olivedrab'
```

```python
plt.plot(E_250Nel_2p_PCQED[0,:], E_250Nel_2p_PCQED[1,:],color=COLOUR1, marker="s", ms="9", mfc="none", label="pPF(250,2)")
plt.plot(E_250Nel_6p_PCQED[0,:],E_250Nel_6p_PCQED[1,:], color=COLOUR3, marker="s", ms="7", mfc="none", label='pPF(250,6)')
plt.plot(E_250Nel_2p_QEDCI[0,:], E_250Nel_2p_QEDCI[1,:],color=COLOUR1, marker="o", ms="6", label="PF-FCI-2/6-311G")
plt.plot(E_250Nel_6p_QEDCI[0,:],E_250Nel_6p_QEDCI[1,:], color=COLOUR3, marker="o", ms="3", label='PF-FCI-6/6-311G')
# plt.plot(E_array[0,:],E_array[5,:])
         
plt.xlabel("Bondlength (Angstroms)")
plt.ylabel("Energy (Hartrees)")
# plt.xlim(1.4, 2.2)
# plt.ylim(-7.9970, -7.9966)
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 12
plt.gca().yaxis.set_major_formatter('{:.3f}'.format)
plt.legend()
plt.tight_layout()
plt.savefig("LiH.png", dpi=150)
plt.show()
```
